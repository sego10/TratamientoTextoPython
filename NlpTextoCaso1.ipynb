```python
ListaMensaje = mensaje.splitlines()
ListaMensaje
```

```python
Resultado=[]
for linea in ListaMensaje: 
  if linea[0:1].isalpha():
    elemento = linea.replace("-",",").replace(" ","")
    elemento = elemento.split(",")  
    nombre = elemento[0]
    id_cliente = elemento[1][0:5]
    id_canal = elemento[1][5:8]
    id_evento = elemento[1][8:13]
    precio = elemento[2]
    Resultado.append([nombre,id_cliente,id_canal,id_evento,precio])

Resultado
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(Resultado,columns=['nombre','cliente','canal','evento','precio'])
df
```
