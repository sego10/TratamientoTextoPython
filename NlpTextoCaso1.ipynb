# Funciones

```python
import pandas as pd

def limpieza(msg):
    ListaMensaje = msg.splitlines()
    Resultado = []
    for linea in ListaMensaje: 
      if linea[0:1].isalpha():
        elemento = linea.replace("-",",")
        elemento = elemento.split(",")  
        nombre = elemento[0]
        id_cliente = elemento[1].replace(" ","")
        id_cliente = id_cliente[0:5]
        id_canal = elemento[1][6:9]
        id_evento = elemento[1][9:14]
        precio = elemento[2]
        Resultado.append([nombre,id_cliente,id_canal,id_evento,precio])
    df = pd.DataFrame(Resultado,columns=['nombre','id','canal','evento','precio'])
    return df

def ObtenerListaIds(df):
  Ids=df['id'].unique()
  Ids.tolist()
  return Ids


def ArreglarId(df,Ids):
  for i in range(Ids.size): 
    df1=df[df["id"]==Ids[i]]
    #print(Ids[i])
    #print(df1)
    seleccion = df1["nombre"].value_counts().idxmax()

    #print(seleccion)
    df.loc[df['id']==Ids[i],"nombre"] = seleccion
  return df
```

```python
#Llamar funciones
dataframe = limpieza(mensaje)
ListaId = ObtenerListaIds(dataframe)
ArreglarId(dataframe,ListaId)
dataframe.sort_values('id')
```
